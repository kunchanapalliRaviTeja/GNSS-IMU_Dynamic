```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from Excel files
file1_data = pd.read_excel('/Users/ravitejakunchanapalli/Downloads/Book1.xlsx')
# Plot comparison of longitudes
plt.plot(file1_data['Lon STD'], label='RTK')
plt.plot(file1_data['Lon STD PPP'], label='PPP')

plt.xlabel('Data Point')
plt.ylabel('Longitude')
plt.title('Comparison of Longitudes')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from Excel files
file1_data = pd.read_excel('/Users/ravitejakunchanapalli/Downloads/Book1.xlsx')

# Extract min and max values of longitudes
min_longitude_rtk = file1_data['Lon STD'].min()
max_longitude_rtk = file1_data['Lon STD'].max()
min_longitude_ppp = file1_data['Lon STD PPP'].min()
max_longitude_ppp = file1_data['Lon STD PPP'].max()

# Plot comparison of longitudes
plt.plot(file1_data['Lon STD'], label=f'RTK (Min: {min_longitude_rtk}, Max: {max_longitude_rtk})')
plt.plot(file1_data['Lon STD PPP'], label=f'PPP (Min: {min_longitude_ppp}, Max: {max_longitude_ppp})')

plt.xlabel('Data Point')
plt.ylabel('Longitude')
plt.title('Comparison of Longitudes')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from Excel files
file1_data = pd.read_excel('/Users/ravitejakunchanapalli/Downloads/Book1.xlsx')

# Plot comparison of Latitudes and Longitudes
plt.scatter(file1_data['Lat STD'], file1_data['Lon STD'], label='RTK', marker='o')
plt.scatter(file1_data['Lat STD PPP'], file1_data['Lon STD PPP'], label='PPP', marker='x')

plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Comparison of Latitudes and Longitudes (RTK vs PPP)')
plt.legend()
plt.show()
```
