```python
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point
import xml.etree.ElementTree as ET

def extract_data_with_timestamp(kml_file):
    tree = ET.parse(kml_file)
    root = tree.getroot()

    data_list = []

    for placemark in root.findall(".//{http://www.opengis.net/kml/2.2}Placemark"):
        latitude = float(placemark.find(".//{http://www.opengis.net/kml/2.2}Data[@name='Latitude']/{http://www.opengis.net/kml/2.2}value").text)
        longitude = float(placemark.find(".//{http://www.opengis.net/kml/2.2}Data[@name='Longitude']/{http://www.opengis.net/kml/2.2}value").text)

        # Replace both dot and comma with an empty string and convert to float
        timestamp_str = placemark.find(".//{http://www.opengis.net/kml/2.2}Data[@name='TimeStamp']/{http://www.opengis.net/kml/2.2}value").text
        timestamp = float(timestamp_str.replace('.', '').replace(',', '.'))

        data_list.append({'latitude': latitude, 'longitude': longitude, 'timestamp': timestamp})

    return data_list

if __name__ == "__main__":
    kml_file_path1 = "/Users/ravitejakunchanapalli/Downloads/PPP DMHR22440044Y_2024-02-02_17-32-22.KML"
    kml_file_path2 = "/Users/ravitejakunchanapalli/Downloads/NMUS22500010A_2024-02-02_11-57-57.KML"

    extracted_data1 = extract_data_with_timestamp(kml_file_path1)
    extracted_data2 = extract_data_with_timestamp(kml_file_path2)

    # Create GeoDataFrames for both files
    geometry1 = [Point(data['longitude'], data['latitude']) for data in extracted_data1]
    gdf1 = gpd.GeoDataFrame(geometry=geometry1, crs="EPSG:4326", data=extracted_data1)

    geometry2 = [Point(data['longitude'], data['latitude']) for data in extracted_data2]
    gdf2 = gpd.GeoDataFrame(geometry=geometry2, crs="EPSG:4326", data=extracted_data2)

    # Plot GeoDataFrames using matplotlib
    fig, ax = plt.subplots()
    gdf1.plot(ax=ax, marker='o', color='blue', markersize=10, alpha=0.7, label='PPP')
    gdf2.plot(ax=ax, marker='o', color='green', markersize=10, alpha=0.7, label='DHMR')

    # Annotate points with timestamps
    for data in extracted_data1:
        ax.annotate(f"{data['timestamp']} (File 1)", (data['longitude'], data['latitude']))

    for data in extracted_data2:
        ax.annotate(f"{data['timestamp']} (File 2)", (data['longitude'], data['latitude']))

    # Set axis labels
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')

    # Show the legend
    ax.legend()

    # Show the plot
    plt.show()
```
